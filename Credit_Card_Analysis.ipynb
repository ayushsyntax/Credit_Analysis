```python
import pandas as pd

def clean_currency_column(column):
    # Remove any non-numeric characters except for the decimal point
    cleaned_column = column.str.replace(r'[^\d.]', '', regex=True)
    # Convert to numeric, setting invalid parsing as NaN
    return pd.to_numeric(cleaned_column, errors='coerce')

# Apply cleaning function to each column
data['Annual_Income'] = clean_currency_column(data['Annual_Income'])
data['Outstanding_Debt'] = clean_currency_column(data['Outstanding_Debt'])
data['Monthly_Balance'] = clean_currency_column(data['Monthly_Balance'])
data['Amount_invested_monthly'] = clean_currency_column(data['Amount_invested_monthly'])

#COULD NOT PERFORM AS NEED TO CONVERT ALL TO STRING THEN USE STRING FUNCTION AND LATER USE TO CONVERT AS PER RWQUIRE
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define the bin range from 0 to 101 with a suitable number of bins
bin_range = range(0, 101, 5)  # Bins from 0 to 101, with a step of 5 years

plt.figure(figsize=(12, 7))  # Larger figure size for better readability
sns.histplot(data['Age'], bins=bin_range, kde=False, color='skyblue', edgecolor='black')  # Histogram with specified bins

plt.title('Distribution of Age of Customers', fontsize=16, weight='bold')  # Title with enhanced font size and weight
plt.xlabel('Age (Years)', fontsize=14)  # X-axis label with enhanced font size
plt.ylabel('Frequency', fontsize=14)  # Y-axis label with enhanced font size
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add grid lines only for y-axis

plt.show()
```

likely balancing career advancement and family responsibilities. To optimize engagement, 
we should tailor our marketing and product offerings to address their specific needs, such as family-related benefits and career growth opportunities. 
Additionally, we can adjust credit limits and risk management strategies to align with the financial stability and
spending patterns typical of this age group.
